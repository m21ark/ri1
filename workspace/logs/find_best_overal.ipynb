```python
import pandas as pd

# Load the CSV file
df = pd.read_csv("filtered_output.csv")

# Filter groups by `kd` and `kp` where both `robot1` and `robot2` have filled data
sum_correct_percentage = df.groupby(['kd', 'kp'])['correct_percentage_distance_wall'].sum().reset_index()

# get max value
max_value = sum_correct_percentage['correct_percentage_distance_wall'].max()

# get the row with the max value
max_row = sum_correct_percentage[sum_correct_percentage['correct_percentage_distance_wall'] == max_value]

print(max_row)
```
